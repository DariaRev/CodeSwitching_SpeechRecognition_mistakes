```python
def make_transcriptions(audio_paths):
    model = SpeechRecognitionModel("jonatasgrosman/wav2vec2-large-xlsr-53-russian")
    #print(len(audio_paths))
    transcriptions = model.transcribe(audio_paths)
    return transcriptions
```

```python
def make_lists(transcriptions):
    pred_value = []
    for el in transcriptions:
        for key, value in el.items():
            if key == 'transcription':
            #print(value)
                pred_value.append(value)
    return pred_value
```

```python
pred_value = []
with open('file_2000_out.txt') as f:
    read = f.readlines()
    for elem in read:
        pred_value.append(elem)
```

```python
def make_dict_for_audios(audio_paths, pred_value):
    di_audios2 = {}
    numb_sent = []
    for n in audio_paths:
        numb_sent.append(n.split('_')[-1])
    for i in range(len(pred_value)):
        di_audios2[numb_sent[i]] = pred_value[i]
    return di_audios2
```

```python
def make_lists_for_align(di_audios2, pred_value, numb):
    with open('file_habr_all.txt') as f:
        read = f.read().split(' . ')
        #print(read)
        #print(len(pred_value))
        read_new = list(map(lambda x: x.replace('\n', ' '), read))
        with open(f'file_for_align_{numb}.txt', 'w') as f1:
            for i in range(len(pred_value)):
                    num = str(list(di_audios2.keys())[i])
                    numb = int(num.split('.')[0])
                    print(numb, pred_value[i], read_new[numb])
                    f1.write(pred_value[i])
                    f1.write(' ||| ')
                    f1.write(read_new[numb])
                    f1.write('\n')
```
